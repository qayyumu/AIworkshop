```python
# =============================================================================
# SECTION RAG CREWAI
# =============================================================================
from crewai import Agent, Task, Crew
# from crewai_tools import ScrapeWebsiteTool
from crewai_tools import RagTool

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

os.environ["CREWAI_TOOLS_ALLOW_UNSAFE_PATHS"] = "true"

# Create a RAG tool with default settings
rag_tool = RagTool()

# Add content from a file
rag_tool.add(data_type="file", path="../Session-03/rag_document.txt")
# Add content from a file
# rag_tool.add(data_type="file", path="Course_content.pdf")

llm="gpt-4o-mini"
```

```python
print("\n" + "="*60)
print("L5: AGENT RAG")
print("="*60)

customer="GenAI"

# Create agents
support_agent = Agent(role="Senior Support Representative",
    goal="Be the most friendly and helpful support representative in your team",
    backstory=f"You work as AI support represebtative and are now working on providing support to client for area {customer}, "
                f"a super important domain for your company. "
                "You need to research the domain to provide the best area of research "
                "Make sure to provide full complete answers, and make no assumptions.",
    allow_delegation=False,
    llm=llm,
    tools=[rag_tool],
    verbose=True
)



inquiry_resolution = Task(description=f"A agent to research and provide the best research article",
    expected_output="A final, detailed, and informative response ready to be sent to the customer.",
    agent=support_agent,
)

# Create crew with memory
crew = Crew(
    agents=[support_agent ],
    tasks=[inquiry_resolution],
    verbose=1,
    memory=False
)

result = await crew.akickoff()
print(result)
```
